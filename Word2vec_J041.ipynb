```
import itertools
import pandas as pd
import string
import nltk
from gensim import models
from nltk import word_tokenize
from nltk.corpus import stopwords
from gensim.models import Word2Vec
from itertools import combinations
from itertools import permutations
from string import digits
```

```
data = pd.read_csv("https://raw.githubusercontent.com/zfz/twitter_corpus/master/full-corpus.csv")
```

```
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
stop_words
```

```
def unique_words(text):
  text = text.translate(text.maketrans('','',string.punctuation))
  text = text.translate(text.maketrans('','',digits))
  tokens = word_tokenize(text)
  result = [word for word in tokens if word not in stop_words]
  return result
```

```
nltk.download('punkt')
data['Tweet_unique_words']= data['TweetText'].apply(unique_words)
```

```
data['Tweet_unique_words'][1]
```

```
word_vec = Word2Vec(data['Tweet_unique_words'], min_count=1)
words = list(word_vec.wv.vocab)
words
```

```
words[:30]
```

```
similar_words = word_vec.wv.most_similar('apps')
```
